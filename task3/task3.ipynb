```python
# 6.1.3 使用分隔格式
```

```python
cat examples/ex7.csv
```

```python
# 对于任何带有单字符分隔符的文件，可以使用python的内建csv模块；
# 要使用它，需要将任一打开的文件或文件型对象传给csv.reader
import csv
f = open('examples/ex7.csv')
reader = csv.reader(f)
# 像遍历文件那样遍历reader会产生元组，元组的值为删除了引号的字符
for line in reader:
    print(line)
```

```python
# 首先将文件读取为行的列表
with open ('examples/ex7.csv') as f:
    lines = list(csv.reader(f))
# 将数据拆分为列名行和数据行
header, values = lines[0], lines[1:]
# 使用字典推导式和表达式zip(*values)生成一个包含数据列的字典，字典中行转置成列：
data_dict = {h: v for h,v in zip(header, zip(*values))}
data_dict
```

```python
# csv有多种不同的风格。如需根据不同的分隔符、字符串引用约定或行钟志峰定义一种新的格式时，
# 可以使用csv.Dialect定义一个简单的子类
class my_dialect(csv.Dialect):
    lineterminator = '\n'
    delimiter = ';'
    quotechar = '"'
    quoting = csv.QUOTE_MINIMAL
reader = csv.reader(f, dialect=my_dialect)

# 或者也可不必定义子类，直接将csv方言参数传入csv.reader的关键字参数：
reader = csv.reader(f, delimiter='|')
```

```python
# csv.Dialect的一些属性及用途
Image(filename = '表6-3_csv方言选项.jpg', width=500, height=300)
```

```python
# 对于复杂的多字符分隔符的文件，无法直接使用csv模块，
# 可以先使用split()或正则re.split()进行拆分，再进行处理
```

```python
# 手动写入被分隔的文件时，使用csv.writer，整个函数和csv.reader的使用类似
with open ('mydata.csv', 'w') as f:
    writer = csv.writer(f, dialect=my_dialect)
    writer.writerow(('one', 'two', 'three'))
    writer.writerow(('1', '2', '3'))
    writer.writerow(('4', '5', '6'))
    writer.writerow(('7', '8', '9'))
```

```python
# 6.2.1 使用HDF5格式
```

```python
frame = pd.DataFrame({'a': np.random.randn(100)})
store = pd.HDFStore('mydata.h5')
store['obj1'] = frame
store['obj1_col'] = frame['a']
store
```

```python
store['obj1']
```

```python
# HDFStore支持2种存储格式，'fixed'和'table'，后者速读更慢，但支持特殊的查询
store.put('obj2', frame, format='table')
store.select('obj2', where=['index >= 10 and index <= 15'])
```

```python
# pandas.read_hdf函数时这些工具的快捷方法
frame.to_hdf('mydata.h5', 'obj3', format='table')
pd.read_hdf('mydata.h5', 'obj3', where=['index< 5'])
```
